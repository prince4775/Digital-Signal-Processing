```python
import librosa
import sounddevice as sd
import numpy as np

# 1. Load your voice recording
filename = r"C:\Users\M Ayub\Downloads\ayub.ogg"
y, sr = librosa.load(filename, sr=None)

print("Original:", y.shape, "Sample rate:", sr)

# 2. Change voice characteristics
#    (positive steps_up = higher pitch, faster voice)
steps_up = 0     # semitones (try -4 for deeper voice)
speed_factor = 1.1  # >1 faster, <1 slower

# Shift pitch
y_shifted = librosa.effects.pitch_shift(y, sr=sr, n_steps=steps_up)
# Change speed
y_stretched = librosa.effects.time_stretch(y_shifted, rate=speed_factor)

# 3. Play original and modified voices
print("Playing original voice...")
sd.play(y, sr)
sd.wait()

print("Playing modified voice...")
sd.play(y_stretched, sr)
sd.wait()

# 4. (Optional) Save the modified audio
import soundfile as sf
sf.write("voice_changed.wav", y_stretched, sr)
print("Saved as voice_changed.wav")
```
